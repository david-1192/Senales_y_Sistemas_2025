# Transformada de Fourier

 - Se presentan los conceptos básicos de la transformada de Fourier, en sus versiones continua, tiempo discreto y transformada discreta (extendida a su versión rápida).

- La transformada de Fourier se puede entender como la extensión de la serie de Fourier para $t\in(-\infty,+\infty)$:

\begin{equation}
X(\omega) = \mathscr{F} \{x(t) \} = \int_{-\infty}^{+\infty}{x(t)e^{-j\omega t}dt}.
\end{equation}

- La respectiva transformada inversa (extensión de la reconstrucción mediante serie de Fourier con $t\in(-\infty,+\infty)$) se define como:

\begin{equation}
x(t) = \mathscr{F}^{-1} \{X(\omega) \} = \frac{1}{2\pi}\int_{-\infty}^{+\infty}{X(\omega)e^{j\omega t}d\omega}.
\end{equation}

- Cabe resaltar entonces que el concepto de espectro en frecuencia para la transformada de Fourier asume que $\omega \in (-\infty,+\infty).$

- En resumen, la serie de Fourier analiza $x(t)\in \mathbb{R},\mathbb{C}$ en tiempo continuo sobre segmento finito $t\in [-T/2,T/2]$ o $t\in [t_i,t_f]$, en espectro en frecuencia discreto $c_n \in \mathbb{R},\mathbb{C}$, con $n\in \mathbb{Z}.$

- La transformada de Fourier analiza $x(t)\in \mathbb{R},\mathbb{C}$ en tiempo continuo sobre segmento infinito de tiempo $t\in (-\infty,+\infty)$, en espectro en frecuencia continuo (función de densidad espectral) $X(\omega)\in \mathbb{R},\mathbb{C}$, con $\omega\in (-\infty,+\infty).$


- Las propiedades principales de la transformada de Fourier son útiles para analizar el espectro en fenómenos físicos clásicos (Ver material apoyo [propiedades de la transformada de Fourier](https://github.com/amalvarezme/SenalesSistemas/blob/master/3_SerieyTransformadaFourier/Propiedades_TransformadaFourier.pdf)).

- Además, algunas de las transformadas básicas se pueden encontrar en las [Tablas de la transformada de Fourier](https://github.com/amalvarezme/SenalesSistemas/blob/master/3_SerieyTransformadaFourier/Tablas_TransformadaFourier.pdf).

# Transformada de Fourier en tiempo discreto

- A partir de la transformada de Fourier, se puede analizar también una señal en tiempo discreto $x[k]\in \mathbb{R} ,\mathbb{C},$ con $k\in\mathbb{Z}$, con el fin de calcular su espectro en frecuencia continuo (función de densidad espectral) $X(\omega)\in \mathbb{R} ,\mathbb{C}$, pero con $\omega$ continua en el intervalo $\omega\in [-\pi,+\pi]$ o $\omega\in [0,2\pi],$ dado que las bases de representación requieren restringir la codificación de la señal $x[k]$ utilizando funciones seno y coseno originales según la expresión:

\begin{equation}
X(\omega) = \sum_{k\in\mathbb{Z}}{x[k]e^{-j\omega k}},
\end{equation}

- **Nota**: Recuerde que cos/sen en tiempo discreto presentan periodicidad de $2\pi$ en la frecuencia angular $\cos[(\omega\pm r 2\pi)k]=\cos[\omega k]$; con $r,k \in \mathbb{Z}$.

- A la expresión anterior se le conoce comunmente como la trasformada de Fouirer en tiempo discreto (Discrete time Fourier transform - DTFT).
y su respectiva transformada inversa (iDTFT) se expresa como:


\begin{equation}
x[k] = \frac{1}{2\pi}\int_{-\pi}^{+\pi}{X(\omega)e^{j\omega k}d\omega}=\frac{1}{2\pi}\int_{0}^{+2\pi}{X(\omega)e^{j\omega k}d\omega}
\end{equation}


# Transformada discreta de Fourier y transformada rápida de Fourier

- Sin embargo, tanto para la serie de Fourier, como para la transformada (continua y discreta en tiempo), se require de la resolución analítica de integrales; además, de la suposición de intervalos infinitos en tiempo y frecuencia (para el caso de la transformada de Fourier), los cuales no son viables ni implementables en la práctica.

- Por consiguiente, el concepto de espectro en frecuencia se extiende a sistemas de proceso de digital (tiempo discreto y  frecuencia discreta), mediante el cálculo de la transformada discreta de Fourier (Discrete Fourier Transform - DFT).

- En este caso, la transformada discreta se define como:

\begin{equation}
X[n] = \sum_{k=0}^{K-1}{x[k]e^{-jk\frac{2\pi n}{K}}},
\end{equation}

**Nota**: observe que el ángulo en la función euler $\theta_n = \frac{k 2\pi n}{K}$se restringe al intervalo $[0,2\pi)$ para $k\in\{0,1,\dots,K-1\}$, dado que se trabaja con funciones cos/sen en tiempo discreto.

- La DFT es una discretización en frecuencia de la DTFT:

\begin{equation}
X[n] = X(\omega)\Big|_{\omega=\frac{2\pi n}{K}},
\end{equation}

- Respecto a la serie de Fourier, la DFT se puede relacionar mediante:

\begin{equation}
X[n] = K c_n,
\end{equation}
asumiendo señal en tiempo discreto y $\omega_o = \frac{2\pi}{K}$ en representación por serie.

- Por otra parte, la transforma discreta de Fourier inversa (iDFT) se define como:

\begin{equation}
x[k] = \sum_{n=0}^{K-1}{X[n]e^{jk\frac{2\pi n}{K}}}.
\end{equation}

Nuevamente, note que las variables tiempo y frecuencia discretas $n,k\in\{0,1,\dots,K-1\}.$ Dado que la frecuencia fundamental se restringe como $\omega_o=\frac{2\pi}{K},$ las bases de representación en tiempo discreto codificaran frecuencias angulares en el rango $[0,2\pi)$ (cosenos y senos originales).

- No obstante, en la práctica el cálculo de las sumatorias es costoso computacionalmente, a medida que $K$ aumenta. Por consiguiente, se cuentan con paquetes computacionales que optimizan el cálculo de la DFT de forma rápida, lo que conlleva a la comunmente conocida transformada rápida de Fourier ([Fast Fourier Transform - FFT](https://es.wikipedia.org/wiki/Transformada_r%C3%A1pida_de_Fourier)).

# Ejemplo:

- Se presenta el cálculo del espectro mediante DFT (`fft` de numpy) para la señal:


\begin{equation}
x(t) = \begin{cases} A, & -T/2 \leq t < 0 \\ -A, & 0 \leq t < T/2 \end{cases}
\end{equation}

```python
# Para graficar
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")

mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)

A = 1 # amplitud de la señal
T = 5 # segundos de gráfica
Fs = 1000 # frecuencia de muestreo definida
tv = np.arange(-T/2,T/2,1/Fs) #vector de tiempo generado
Nm = len(tv) #numero de muestras requeridas

#se crea la señal A y -A concatenando
xo = np.concatenate((A*np.ones(int(Nm/2)), -A*np.ones(int(Nm/2))), axis=0)
#graficar señal original
plt.plot(tv,xo,'r',linewidth=4)
plt.grid()
plt.xlabel("t[s]",fontsize = 14)
plt.ylabel("x(t)",fontsize = 14)
plt.show()
```

```python
#cálculo de la FFT de la señal signo invertido
vf = np.fft.fftfreq(len(xo),1/Fs) #calculo vector de frecuencias
Xw = np.fft.fft(xo) #cálculo FFT

plt.stem(vf,abs(Xw),'r')
plt.title(r'|X[n]|')
plt.xlabel(r'$f$[Hz]',fontsize = 14)
plt.ylabel(r'$|X[n]|$',fontsize = 14)
plt.axis('tight')
plt.grid()
plt.show()
plt.stem(vf,abs(Xw)/len(vf),'r')
plt.title(r'|X[n]|/K=c_n')
plt.xlabel(r'$f$[Hz]',fontsize = 14)
plt.ylabel(r'$|X[n]|/K$',fontsize = 14)
plt.show()
```

**Nota**: La función `np.fft.fftfreq` cálcula el equivalente en Hz de cada armónico, teniendo en cuenta que en tiempo discreto:

$$\theta_n = \frac{k 2\pi n}{K} = k \omega_n = k n \omega_o$$

Además, en el proceso de discretización ($F_s$ frecuencia de muestreo):

$$\omega_n t = 2\pi F_n t =  \frac{2\pi F_n k}{F_s} = 2 \pi f n k = \frac{2 \pi n k}{K}$$

donde $f=\frac{1}{K}$ es equivalente a la frecuencia en Hz en tiempo discreto.

Así:

$$\frac{2\pi F_n k}{F_s}=\frac{2 \pi n k}{K}$$

y:

$$F_n = \frac{F_s n}{K} [Hz]$$




- Note que:

$x(t) = \frac{1}{2\pi}\int_{-\infty}^\infty X(\omega) e^{j\omega t} d\omega$

$x^*(t) = \left( \frac{1}{2\pi}\int_{-\infty}^\infty X(\omega) e^{j\omega t} d\omega\right)^*=\frac{1}{2\pi}\int_{-\infty}^\infty X^*(\omega) e^{-j\omega t} d\omega$

Haciendo $\omega' = - \omega$:


$x^*(t) = \frac{1}{2\pi}\int_{-\infty}^\infty X^*(-\omega') e^{j\omega' t} d\omega'$

Entonces:

$\mathscr{F}\{x^*(t)\} = X^*(-\omega) = \mathscr{Re}\{X(-\omega) \} -j\mathscr{Im}\{X(-\omega) \} $

Y

$\mathscr{F}\{x(t)\} = X(\omega) = \mathscr{Re}\{X(\omega) \} +j\mathscr{Im}\{X(\omega) \} $

Por lo tanto:

$\mathscr{Re}\{X(\omega) \}=\mathscr{Re}\{X(-\omega) \}$
$\mathscr{Im}\{X(\omega) \}=-\mathscr{Im}\{X(-\omega) \}$

-Dado que el espectro es simétrico para señales reales ($x(t)\in \mathbb{R}$), se sugiere utilizar la expresión simplificada `rfft`:

```python
vf = np.fft.rfftfreq(len(xo),1/Fs) #calculo vector de frecuencias
Xw = np.fft.rfft(xo) #cálculo FFT
plt.stem(vf,abs(Xw),'r')
plt.title(r'|X[n]|')
plt.xlabel(r'$f$[Hz]',fontsize = 14)
plt.ylabel(r'$|X[n]|$',fontsize = 14)
plt.axis('tight')
plt.grid()
plt.show()
plt.stem(vf,abs(Xw)/(2*len(vf)),'r')
plt.title(r'|X[n]|/K=c_n')
plt.xlabel(r'$f$[Hz]',fontsize = 14)
plt.ylabel(r'$|X[n]|/K$',fontsize = 14)
plt.grid()
plt.show()
```

- La reconstrucción de x[k] se puede obtener mediante la transformada inversa:

```python
xe = np.fft.irfft(Xw) #transformada inversa
plt.plot(tv,xe)
plt.title(r'Reconstrucción completa')
plt.xlabel(r't',fontsize = 14)
plt.ylabel(r'$x[n]$',fontsize = 14)
plt.show()
```

- Se pueden escoger o ponderar los coeficientes para filtrar (alterar) la reconstrucción de la señal.

```python
Xwf = Xw.copy()
Xwf[10:] = 0 #solo se tienen en cuenta los 10 primeros armónicos
xe2 = np.fft.irfft(Xwf)
plt.plot(tv,xe2)
plt.show()
```

**EJERCICIO 1:**

En este ejercicio se presentará y describirá el modelo que utiliza la FFT para calcular los armónicos.

La Transformada Rápida de Fourier (FFT) es un algoritmo computacionalmente eficiente para calcular la Transformada Discreta de Fourier (DFT) de una secuencia. La DFT, a su vez, descompone una señal discreta en el dominio del tiempo en sus componentes de frecuencia discreta, también conocidos como armónicos.

El modelo fundamental que subyace a la DFT (y por lo tanto a la FFT) se basa en la representación de una señal discreta $x[k]$ como una suma ponderada de exponenciales complejas discretas (análogas a las senoides y cosenoides en tiempo continuo). La fórmula de la DFT para una señal $x[k]$ de longitud $K$ está dada por:

$$
X[n] = \sum_{k=0}^{K-1}{x[k]e^{-jk\frac{2\pi n}{K}}}, \quad \text{para } n = 0, 1, \dots, K-1
$$

Donde:
- $x[k]$ es la señal de entrada en el dominio del tiempo discreto, con $k$ representando el índice de la muestra ($k=0, 1, \dots, K-1$).
- $X[n]$ es el espectro de frecuencia discreta, con $n$ representando el índice del armónico o componente de frecuencia ($n=0, 1, \dots, K-1$).

- $e^{-jk\frac{2\pi n}{K}}$ son las bases de representación, que son exponenciales complejas discretas. El término $\frac{2\pi n}{K}$ representa la frecuencia angular discreta asociada al n-ésimo armónico. Este término se restringe al intervalo $[0, 2\pi)$ porque las funciones seno y coseno en tiempo discreto tienen una periodicidad de $2\pi$ en la frecuencia angular.

Cada valor $X[n]$ representa la "cantidad" (magnitud y fase) de la componente de frecuencia (armónico) con frecuencia angular $\omega_n = \frac{2\pi n}{K}$ (o frecuencia lineal $F_n = \frac{F_s n}{K}$ en Hz, donde $F_s$ es la frecuencia de muestreo) presente en la señal $x[k]$.

El modelo de la FFT (específicamente los algoritmos como el de Cooley-Tukey) no calcula directamente la suma de la DFT para cada $n$ de forma independiente (lo que tendría una complejidad $O(K^2)$), sino que explota las propiedades de simetría y periodicidad de las exponenciales complejas para descomponer el cálculo de una DFT de tamaño $K$ en DFTs más pequeñas, reduciendo significativamente la complejidad a $O(K \log_2 K)$. Esto permite calcular eficientemente los armónicos incluso para señales de gran longitud.

El modelo subyacente a la FFT para el cálculo de armónicos es la descomposición de una señal discreta en el tiempo en una suma ponderada de exponenciales complejas discretas, donde cada peso ($X[n]$) representa la contribución de un armónico específico a la señal original. La FFT es simplemente un algoritmo eficiente para realizar esta descomposición (la DFT).

**EJERCICIO 2:**

En este punto se busca Realizar una comparación del tiempo de cómputo requerido para el cálculo de la transformada discreta de Fourier utilizado la estimación por sumatoria vs la FFT. Probaremos distintos valores de longitud de la señal de entrada (4,8,16,32,64,128,256,512,1024,2048,4096). Además, se presentará una gráfica comparativa de los tiempos de ejecución en segundos vs la longitud de la señal de entrada.

```python
import numpy as np
import time
import matplotlib.pyplot as plt

# Función para calcular la DFT utilizando la fórmula de la sumatoria
def dft_sumatoria(x):
    """
    Calcula la Transformada Discreta de Fourier (DFT)
    de una señal x utilizando la fórmula de la sumatoria.

    Args:
        x (np.ndarray): La señal de entrada (un array NumPy).

    Returns:
        np.ndarray: El espectro de frecuencia (un array NumPy de números complejos).
    """
    N = len(x)
    X = np.zeros(N, dtype=np.complex128)  # Inicializamos el array de salida con ceros complejos
    for n in range(N):
        for k in range(N):
            # Fórmula de la DFT: X[n] = sum_{k=0}^{N-1} x[k] * exp(-j * 2 * pi * n * k / N)
            X[n] += x[k] * np.exp(-1j * 2 * np.pi * n * k / N)
    return X

# Tamaños de señal para la comparación
signal_lengths = [4, 8, 16, 32, 64, 128, 256, 512, 1024, 2048, 4096]

# Listas para almacenar los tiempos de ejecución
times_sumatoria = []
times_fft = []

# Realizar la comparación para cada tamaño de señal
for N in signal_lengths:
    # Crear una señal de prueba aleatoria
    signal = np.random.rand(N)

    # Medir el tiempo de ejecución de la DFT por sumatoria
    start_time = time.time()
    dft_sumatoria(signal)
    end_time = time.time()
    times_sumatoria.append(end_time - start_time)

    # Medir el tiempo de ejecución de la FFT de NumPy
    start_time = time.time()
    np.fft.fft(signal)
    end_time = time.time()
    times_fft.append(end_time - start_time)

# Graficar los resultados
plt.figure(figsize=(10, 6))
plt.plot(signal_lengths, times_sumatoria, label='DFT por sumatoria', marker='o')
plt.plot(signal_lengths, times_fft, label='FFT (NumPy)', marker='x')
plt.xlabel('Longitud de la señal de entrada')
plt.ylabel('Tiempo de ejecución (segundos)')
plt.title('Comparación del tiempo de ejecución: DFT vs FFT')
plt.xscale('log')  # Usar escala logarítmica en el eje x para mejor visualización
plt.yscale('log')  # Usar escala logarítmica en el eje y para mejor visualización
plt.legend()
plt.grid(True)
plt.show()
```

**Explicación de la Gráfica de Tiempos de Ejecución:**

La gráfica anterior muestra la comparación del tiempo de ejecución en segundos requerido para calcular la Transformada Discreta de Fourier (DFT) utilizando dos métodos diferentes en función de la longitud de la señal de entrada:

1.  **"DFT por sumatoria" (curva naranja con marcadores 'x'):** Esta curva representa el tiempo que tarda en calcularse la DFT utilizando la implementación directa de la fórmula de la sumatoria. Como se puede observar, el tiempo de ejecución aumenta de manera significativa a medida que crece la longitud de la señal. Esto se debe a que la complejidad computacional de este enfoque es cuadrática, O(N^2), donde N es la longitud de la señal. Por cada elemento en la salida de la DFT, se realiza una suma sobre todos los elementos de la señal de entrada.

2.  **"FFT (NumPy)" (curva azul con marcadores 'o'):** Esta curva muestra el tiempo de ejecución de la función `np.fft.fft` de la librería NumPy, que implementa la Transformada Rápida de Fourier (FFT). Es evidente que esta curva crece mucho más lentamente que la de la DFT por sumatoria. Esto se debe a que la FFT es un algoritmo mucho más eficiente, con una complejidad computacional de O(N log N). A medida que la longitud de la señal aumenta, la diferencia en el tiempo de ejecución entre ambos métodos se vuelve drástica, demostrando la superioridad de la FFT para el cálculo eficiente de la DFT en la práctica.

Las escalas logarítmicas en ambos ejes (`Longitud de la señal de entrada` y `Tiempo de ejecución (segundos)`) permiten visualizar de mejor manera la diferencia de crecimiento entre las dos curvas a lo largo de un amplio rango de longitudes de señal. La brecha entre las dos curvas se ensancha considerablemente a medida que nos movemos hacia la derecha en el eje x (mayor longitud de señal).
