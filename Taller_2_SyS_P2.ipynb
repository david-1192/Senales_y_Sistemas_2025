### **2.3** Solucin

Demostracin de si lo siguientes sistemas de la forma y = H{x} son sistemas lineales e invariantes en el tiempo.(SLIT): https://github.com/david-1192/Senales_y_Sistemas_2025/blob/main/2.3_SyS_T2.pdf

###Simulacin del punto 2.3.
###SISTEMA 1
$y[n] = \frac{x[n]}{3} + 2x[n-1] - y[n-1]$ .


```python
#Librerias a utilizar
import numpy as np
import matplotlib.pyplot as plt

#Funcin para crear el arreglo del sistema
def sistema1(x):
    y = np.zeros_like(x)
    for n in range(1, len(x)):
        y[n] = x[n]/3 + 2*x[n-1] - y[n-1]
    return y

# Seal de prueba
n = np.arange(0,30)
x = np.ones_like(n)  # escaln unitario
y = sistema1(x)

print("El sistema 1 es un SLIT")

plt.stem(n, y)
plt.title("Respuesta del Sistema 1")
plt.xlabel("n")
plt.ylabel("y[n]")
plt.grid(True)
plt.show()
```

##SISTEMA 2: Sumatoria no lineal
$y[n] = \sum_{k=-\infty}^{n} x^2[k]$

```python
#Funcin para crear el sistema
def sistema2(x):
    y = np.zeros_like(x)
    for n in range(len(x)):
        y[n] = np.sum(x[:n+1]**2)
    return y

x = np.ones_like(n)
y = sistema2(x)

print ("El sistema 2 NO es un SLIT")

# Grfica
plt.stem(n, y)
plt.title("Respuesta del Sistema 2")
plt.xlabel("n")
plt.ylabel("y[n]")
plt.grid(True)
plt.show()
```

# SISTEMA 3: Funcin mediana sobre ventana de tamao 3
$y[n] = \mathrm{median}(x[n-1],\, x[n],\, x[n+1])$

```python
# Importar las librerias necesarias
from scipy.signal import medfilt
import numpy as np
import matplotlib.pyplot as plt

#Funcion para crear el sistema
def sistema3(x):
    return medfilt(x, kernel_size=3)

x = np.concatenate((np.ones(5), np.zeros(5), np.ones(5)))
y = sistema3(x)

print("El sistema NO es un SLIT")

plt.stem(np.arange(len(x)), y)
plt.title("Respuesta del Sistema 3 (Mediana)")
plt.xlabel("n")
plt.ylabel("y[n]")
plt.grid(True)
plt.show()
```

#SISTEMA 4
$y(t) = A \cdot x(t) + B \qquad \text{con } A, B \in \mathbb{R}$

```python
# Libreras
import numpy as np
import matplotlib.pyplot as plt

# Parmetros del sistema
A = 2
B = 4  # Se Cambia este valor a 0 para que el sistema sea un SLIT

# Seal de entrada: senoidal
t = np.linspace(0, 2*np.pi, 1000)
x = np.sin(t)

# Aplicamos el sistema
y = A * x + B

# Grfica
plt.figure(figsize=(10, 4))
plt.plot(t, x, label='x(t): entrada')
plt.plot(t, y, '--', label='y(t) = Ax(t) + B')
plt.title(f"Sistema: y(t) = {A}x(t) + {B}")
plt.xlabel("t")
plt.ylabel("Amplitud")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Evaluacin SLIT
# La linealidad solo se cumple si B == 0
es_lineal = (B == 0)

# La invarianza se cumple siempre en este sistema (desplazamiento temporal se conserva)
es_invariante = True

# Resultado
if es_lineal and es_invariante:
    print(" El sistema es SLIT (Lineal e Invariante en el tiempo).")
elif not es_lineal and es_invariante:
    print(" El sistema NO es SLIT: NO es lineal pero s es invariante en el tiempo.")
else:
    print(" El sistema NO es SLIT.")

```

### **2.4** Solucin

Hallar la salida y[n] de un SLIT ante la entrada x[n]={15, 5,3, 0, 5, 7,1}, con respuesta al impulso h[n]={1,2, 0, 1,2}, donde n{0,1,2, . . . ,N} y n=0 para x[n].

- Dada la respuesta al impulso en tiempo discreto $h[n]\in\mathbb{R},\mathbb{C}$ de un SLIT $\mathsf{H}\{\cdot\},$ la respuesta del sistema $y[n]\in\mathbb{R},\mathbb{C}$ ante la entrada discreta $x[n]\in\mathbb{R},\mathbb{C}$, se puede calcular mediante el operador convolucin:

$$y[n]=x[n]*h[n] = \sum^\infty_{k=-\infty}x[k]h[n-k]$$


- En general, se cuenta con vectores de tamao finito, y con sistemas causales, por ende, si $x[n]$ cuenta con $N$ elementos y $h[n]$ con $L$ elementos, la salida $y[n]$ tendra, en el mejor de los casos, $M = N + L - 1$ elementos distintos de 0.

Se presenta la simulacin del clculo de la salida de un SLIT en tiempo discreto utilizando la respuesta al impulso y la convolucin discreta implementando punto a punto el proceso de suma y comparando los resultados con la librera [convolve]

**Librerias a Importar y configuracin interactiva**

```python
%matplotlib inline
from ipywidgets import interact #crear graficos interactivos
import numpy as np
import matplotlib.pyplot as plt
```

**Definir seales correctas (conforme al enunciado)**

```python
# Entrada x[n] y respuesta al impulso h[n]
x = np.array([-15, 5, -3, 0, 5, 7, -1]).reshape(-1, 1)  # n=0 est en ndice 2
h = np.array([1, -2, 0, 1, -2]).reshape(-1, 1)          # n=0 est en ndice 2
```

**Funcin convgraf()**

```python
def convgraf(h, x):
    lx = len(x)
    lh = len(h)
    M = lx + lh - 1  # longitud de la salida
    xm = np.zeros((lh, M))  # almacenamiento de x reflejado y desplazado
    ym = np.zeros(M)        # salida final

    # Reflejar x[k] -> x[-k]
    x_flip = np.flip(x, axis=0).flatten()

    for i in range(M):
        # Desplazar x_flip para formar x[k] en n = i
        x_shift = np.zeros(lh)
        for j in range(lh):
            if 0 <= i - j < lx:
                x_shift[j] = x_flip[i - j]
        xm[:, i] = x_shift
        ym[i] = np.dot(x_shift, h.flatten())  # producto punto

    return xm, ym, h

```

**Ejecutar la convolucin y visualizarla**

```python
def plot_conv(k, xm, ym, h):
    plt.figure(figsize=(10, 5))

    plt.stem(xm[:, k], markerfmt='+', label='$x[k]$ desplazado')  # x reflejado y desplazado
    plt.stem(h, linefmt='g', markerfmt='.', label='$h[n-k]$')     # h fija
    plt.stem(ym[:k+1], markerfmt='s', linefmt='r', label='$y[n]$ acumulada')  # salida parcial

    plt.title('Proceso de convolucin: $y[n] = x[n] * h[n]$', fontsize=14)
    plt.xlabel('n (ndice de tiempo discreto)', fontsize=12)
    plt.ylabel('Amplitud', fontsize=12)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()


```

```python
xm, ym, h_used = convgraf(h, x)

@interact(k=(0, len(ym)-1, 1))
def show_frame(k=0):
    plot_conv(k, xm, ym, h_used)
```

**Visualizacin de los valores.**

```python
# Calcular ndice de n=0 en la salida
n0_x = 2  # n=0 en x
n0_h = 2  # n=0 en h
n0_y = n0_x + n0_h  # n=0 en y

# Crear eje temporal para y[n]
n_y = np.arange(-n0_y, len(ym) - n0_y)

# Mostrar resultados
print("Salida completa de la convolucin y[n]:\n")
for n_val, y_val in zip(n_y, ym):
    print(f"n = {n_val:>2} --> y[n] = {y_val:>6.1f}")
```

### Respuesta al escaln

Para esta parte del ejercicio, se analiza la respuesta del sistema \( s[n] \), dada por:


s[n] = {-1,6,-10,3,1,-10,2,5}


donde el valor con el smbolo () en el nmero 3 corresponde al instante ( n = 0 ).  
Este tipo de respuesta puede interpretarse como una **respuesta al escaln unitario**, es decir, cmo responde el sistema cuando la entrada es:


u[n] = \begin{cases}
1, & n \geq 0 \\
0, & n < 0
\end{cases}


La relacin entre la respuesta al escaln \( s[n] \) y la respuesta al impulso \( h[n] \) es:

\[
h[n] = s[n] - s[n-1]
\]

En esta parte, simplemente analizamos cmo la entrada \( x[n] \) excita un sistema con esta respuesta, usando convolucin discreta:  
\[
y[n] = x[n] * s[n]
\]

El resultado de esta convolucin representa la salida del sistema con entrada \( x[n] \) y una **respuesta escalonada**, no impulsiva.


```python
# Nueva respuesta al escaln s[n]
s = np.array([-1, 6, -10, 3, 1, -10, 2, 5]).reshape(-1, 1)  # n = 0 en ndice 3

# Convolucin con la nueva "respuesta al sistema"
def convgraf_step(s, x):
    lx = len(x)
    lh = len(s)
    M = lx + lh - 1
    xm = np.zeros((lh, M))
    ym = np.zeros(M)

    x_flip = np.flip(x, axis=0).flatten()

    for i in range(M):
        x_shift = np.zeros(lh)
        for j in range(lh):
            if 0 <= i - j < lx:
                x_shift[j] = x_flip[i - j]
        xm[:, i] = x_shift
        ym[i] = np.dot(x_shift, s.flatten())

    return xm, ym, s

# Ejecutamos
xm_step, ym_step, s_used = convgraf_step(s, x)

# Visualizacin paso a paso con etiquetas
def plot_step(k, xm, ym, s):
    plt.figure(figsize=(10, 5))
    plt.stem(xm[:, k], markerfmt='+', label='$x[k]$ desplazado')
    plt.stem(s, linefmt='g', markerfmt='.', label='$s[n-k]$')
    plt.stem(ym[:k+1], markerfmt='s', linefmt='r', label='$y[n]$')
    plt.title('Convolucin con respuesta al escaln: $y[n] = x[n] * s[n]$', fontsize=14)
    plt.xlabel('n (ndice de tiempo discreto)', fontsize=12)
    plt.ylabel('Amplitud', fontsize=12)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Widget interactivo
@interact(k=(0, len(ym_step)-1, 1))
def show_step_frame(k=0):
    plot_step(k, xm_step, ym_step, s_used)
```

**Visualizacin de los valores**

```python
# ndice de n=0 en x[n] y s[n]
n0_x = 2
n0_s = 3
n0_y_step = n0_x + n0_s  # ndice de n=0 en la convolucin y[n]

# Eje de tiempo asociado a la salida
n_y_step = np.arange(-n0_y_step, len(ym_step) - n0_y_step)

# Mostrar resultados numricos
print("Salida completa de la convolucin con respuesta al escaln s[n]:\n")
for n_val, y_val in zip(n_y_step, ym_step):
    print(f"n = {n_val:>2} --> y[n] = {y_val:>6.1f}")

```
